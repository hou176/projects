```python

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
x = np.arange(0,  3  * np.pi,  0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
#  subplot  2 1  
#  subplot
plt.subplot(2,  1,  1)  
#  
plt.plot(x, y_sin) 
plt.title('Sine')  
#  subplot 
plt.subplot(2,  1,  1) 
plt.plot(x, y_cos) 
plt.title('Cosine')  
# 
plt.show()
```
