```python
# 进行建模
fenceng_5 = fenceng[['age','jincheng','sex']]
# 使用数据标准化  使用sklearn中预处理的scale
from sklearn.preprocessing import scale
x = pd.DataFrame(scale(fenceng_5))

# 使用cluster建模
from sklearn import  cluster
model = cluster.KMeans(n_clusters=3,random_state=10)
model.fit(x)

# 提取标签  查看分类结果
fenceng_5['cluster'] = model.labels_
fenceng_5.head(35)
```

```python
# 模型评估优化
fenceng_5.groupby(['cluster'])['age'].describe()
```

```python
fenceng_5.groupby(['cluster'])['jincheng'].describe()
```

```python
# 模型评估
from sklearn import  metrics
x_cluster = model.fit_predict(x)
score = metrics.silhouette_score(x,x_cluster)
print(score)
```

```python
model=cluster.KMeans(n_clusters=3,random_state=10)
model.fit(x)
centers=pd.DataFrame(model.cluster_centers_)
centers.to_csv('fenceng_3_3.csv')
```
