```python
import pandas as pd
import datetime
import numpy as np
```

```python
data = pd.read_excel(r'data/2022.04/1.xlsx')
```

```python
data
```

```python
data1 = data[['商品名称','支付金额','商品单价','购买数量','省']]

data1['金额占比'] = data1['支付金额']  / data1['支付金额'].sum() 

data1['金额占比'] = data1['金额占比'].apply(lambda x: format(x, '.6%')) 

data1.to_excel(r'data/2022.04/1.单品全国售卖详情.xlsx')
```
