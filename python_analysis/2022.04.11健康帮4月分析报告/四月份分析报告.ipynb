# 用户活跃（3、4）

### 用户活跃详细数据

```python
sql2 = "SELECT	user_id,group_id,	province,	city,	created_at,	updated_at   ,DATE_FORMAT(created_at,'%Y-%m-%d')  as 'time'  FROM	jdt_user_logon_in WHERE  	updated_at > '2022-04-01 00:00:00'   and  updated_at < '2022-05-01 00:00:00'  GROUP BY	DATE_FORMAT(created_at,'%Y-%m-%d') ,	user_id"
data2 = jdt_db(sql2)
data2['des'] = np.where(data2['updated_at']>data2['created_at'],'重复',np.where(data2['updated_at']==data2['created_at'],'新增','错误'))
data2['time'] = pd.to_datetime(data2['created_at'],format='%Y-%m-%d').dt.date
```

```python
sql3 = "SELECT	user_id,group_id,	province,	city,	created_at,	updated_at   ,DATE_FORMAT(created_at,'%Y-%m-%d')  as 'time'  FROM	jdt_user_logon_in WHERE  	updated_at >= '2022-03-01 00:00:00'  AND updated_at < '2022-04-01 00:00:00'    GROUP BY	DATE_FORMAT(created_at,'%Y-%m-%d') ,user_id"
data3 = jdt_db(sql3)
data3['des'] = np.where(data3['updated_at']>data3['created_at'],'重复',np.where(data3['updated_at']==data3['created_at'],'新增','错误'))
data3['time'] = pd.to_datetime(data3['created_at'],format='%Y-%m-%d').dt.date
```

```python
data_2_3 = pd.concat([data2,data3],axis = 0)
data_2_3.to_csv(r'data/data2.csv')
```

### 用户活跃每日总数据

```python
list_time = ['2022-04-01','2022-04-02','2022-04-03','2022-04-04','2022-04-05','2022-04-06','2022-04-07','2022-04-08','2022-04-09','2022-04-10','2022-04-11','2022-04-12','2022-04-13','2022-04-14','2022-04-15','2022-04-16','2022-04-17','2022-04-18','2022-04-19','2022-04-20','2022-04-21','2022-04-22','2022-04-23','2022-04-24','2022-04-25','2022-04-26','2022-04-27','2022-04-28','2022-04-29','2022-04-30','2022-03-01','2022-03-02','2022-03-03','2022-03-04','2022-03-05','2022-03-06','2022-03-07','2022-03-08','2022-03-09','2022-03-10','2022-03-11','2022-03-12','2022-03-13','2022-03-14','2022-03-15','2022-03-16','2022-03-17','2022-03-18','2022-03-19','2022-03-20','2022-03-21','2022-03-22','2022-03-23','2022-03-24','2022-03-25','2022-03-26','2022-03-27','2022-03-28','2022-03-29','2022-03-30','2022-03-31']

data_2_3['time'] = data_2_3['time'].astype(str)
data_2_3_drop = data_2_3.drop_duplicates(subset='user_id',inplace=False)
huoyue_cishu  = {}
huoyue_renshu  = {}
for i in list_time:
    x = data_2_3[data_2_3['time'] == i]
    y = data_2_3_drop[data_2_3_drop['time'] == i]
    cishu = x['user_id'].count()
    renshu = y['user_id'].count()
    huoyue_cishu[i] = cishu
    huoyue_renshu[i] = renshu

huoyue_cishu = pd.DataFrame.from_dict(huoyue_cishu, orient='index',columns=['活跃次数'])
huoyue_renshu = pd.DataFrame.from_dict(huoyue_renshu, orient='index',columns=['活跃人数'])

huoyue = pd.concat([huoyue_cishu,huoyue_renshu],axis = 1)
huoyue['平均活跃次数'] = huoyue['活跃次数']   / huoyue['活跃人数']
huoyue.to_csv(r'data/data3.csv')
```

# 用户留存(3、4)

```python
sql4 = "SELECT	a.id,	a.user_province,	DATE_FORMAT(FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ), '%Y-%m-%d' )  AS '日期',	FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) AS '时间',	a.age,	a.gender,	a.group_id,	b.created_at,	DATE_FORMAT( b.created_at, '%Y-%m-%d' ) AS 'time' FROM	jdt_user AS a	LEFT JOIN jdt_user_logon_in AS b ON a.id = b.user_id   WHERE	FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) >= '2022-04-01 00:00:00' 	AND FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) < '2022-05-01 00:00:00'"
data4 = jdt_db(sql4)

data4['时间']= pd.to_datetime(data4['时间'])
data4['created_at']= pd.to_datetime(data4['created_at'])
data4['日期']= pd.to_datetime(data4['日期'])
data4['time']= pd.to_datetime(data4['time'])

data4['相隔时间'] = data4['created_at'] - data4['时间']

data4['相隔时间'] = data4['相隔时间'].map(lambda x: x/np.timedelta64(1,'h'))

data4['留存'] = np.where(((data4['相隔时间']>= 0.05  )   & (data4['相隔时间'] <24 )),'次日留存',np.where(((data4['相隔时间']>= 24  )   & (data4['相隔时间'] <186 )),'7日留存',np.where(((data4['相隔时间']>= 186  )   & (data4['相隔时间'] <360 )),'15日留存',np.where(((data4['相隔时间']>= 360 )  & (data4['相隔时间'] <720 )),'30日留存','未知'))))

data4.to_csv(r'data/data4.csv')
```
