```python
from demo  import jdt_db
import pandas as pd
import datetime
import numpy as np
```

```python
sql = "SELECT	id,	nickname,	user_from,user_province,province_id,	p_name,	FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' )  as '时间',	age,	gender,	group_id FROM	jdt_user WHERE	FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) >= '2022-03-01 00:00:00' 	AND FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) < '2022-05-01 00:00:00'"
data1 = jdt_db(sql)

data1['time'] = pd.to_datetime(data1['时间'],format='%Y-%m-%d').dt.month

data1['用户注册方式'] = np.where(data1['p_name']=='官方直播中心','自绑定','被邀请')

data1.to_csv(r'data/data1.csv')
```

```python
sql2 = "SELECT	user_id,group_id,	province,	city,	created_at,	updated_at   ,DATE_FORMAT(created_at,'%Y-%m-%d')  as 'time'  FROM	jdt_user_logon_in WHERE  	updated_at > '2022-04-01 00:00:00' GROUP BY	DATE_FORMAT(created_at,'%Y-%m-%d') ,	user_id"


data2 = jdt_db(sql2)
data2['des'] = np.where(data2['updated_at']>data2['created_at'],'重复',np.where(data2['updated_at']==data2['created_at'],'新增','错误'))

data2['time'] = pd.to_datetime(data2['created_at'],format='%Y-%m-%d').dt.date



data2.to_csv(r'data/data2.csv')
```

```python
sql3 = "SELECT	user_id,group_id,	province,	city,	created_at,	updated_at   ,DATE_FORMAT(created_at,'%Y-%m-%d')  as 'time'  FROM	jdt_user_logon_in WHERE  	updated_at >= '2022-03-01 00:00:00'  AND updated_at < '2022-04-01 00:00:00'    GROUP BY	DATE_FORMAT(created_at,'%Y-%m-%d') ,user_id"




data3 = jdt_db(sql3)
data3['des'] = np.where(data3['updated_at']>data3['created_at'],'重复',np.where(data3['updated_at']==data3['created_at'],'新增','错误'))

data3['time'] = pd.to_datetime(data3['created_at'],format='%Y-%m-%d').dt.date


```

```python
data_2_3 = pd.concat([data2,data3],axis = 0)

data_2_3.to_csv(r'data/data2.csv')
```
