```python
sql_time1 = "'2022-03-01 00:00:00'"
sql_time2 = "'2022-04-01 00:00:00'"

sql = "SELECT	a.id,	FROM_UNIXTIME( a.jointime, '%Y-%m-%d %H:%i:%s' ) AS '加入时间',	a.p_name AS '父级名称',	a.gender AS '性别',	a.age AS '年龄',	a.birthday AS '生日',	a.province AS '地区编码',	a.user_province AS '注册省份',	a.user_city AS '注册城市',	b.conference_id,	b.c AS '最近参加会务时间',	b.d AS '参加会务次数' ,	g.group_id as '用户级别',	g.f as '用户最后登陆时间',g.e as '用户登录次数'  FROM	jdt_user AS a	LEFT JOIN (	SELECT		conference_id,		buyer_id,		MAX( created_at ) AS c,		count( buyer_id ) AS d 	FROM		jdt_conference_order 	WHERE		created_at >=" + sql_time1 + "	AND created_at < " + sql_time2 + " 	GROUP BY buyer_id) AS b ON b.buyer_id = a.id LEFT JOIN (SELECT user_id,group_id,MAX(created_at) as f, count(user_id) as e FROM jdt_user_logon_in WHERE created_at >= " + sql_time1 +  "	AND created_at < "+ sql_time2 +  	"GROUP BY	user_id) as g on g.user_id = a.id  WHERE FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) >= "+ sql_time1 + "	AND FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) <" +sql_time2
data1 = jdt_db(sql)
```

```python
sql_time1 = "'2022-02-01 00:00:00'"
sql_time2 = "'2022-03-01 00:00:00'"

sql = "SELECT	a.id,	FROM_UNIXTIME( a.jointime, '%Y-%m-%d %H:%i:%s' ) AS '加入时间',	a.p_name AS '父级名称',	a.gender AS '性别',	a.age AS '年龄',	a.birthday AS '生日',	a.province AS '地区编码',	a.user_province AS '注册省份',	a.user_city AS '注册城市',	b.conference_id,	b.c AS '最近参加会务时间',	b.d AS '参加会务次数' ,	g.group_id as '用户级别',	g.f as '用户最后登陆时间',g.e as '用户登录次数'  FROM	jdt_user AS a	LEFT JOIN (	SELECT		conference_id,		buyer_id,		MAX( created_at ) AS c,		count( buyer_id ) AS d 	FROM		jdt_conference_order 	WHERE		created_at >=" + sql_time1 + "	AND created_at < " + sql_time2 + " 	GROUP BY buyer_id) AS b ON b.buyer_id = a.id LEFT JOIN (SELECT user_id,group_id,MAX(created_at) as f, count(user_id) as e FROM jdt_user_logon_in WHERE created_at >= " + sql_time1 +  "	AND created_at < "+ sql_time2 +  	"GROUP BY	user_id) as g on g.user_id = a.id  WHERE FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) >= "+ sql_time1 + "	AND FROM_UNIXTIME( jointime, '%Y-%m-%d %H:%i:%s' ) <" +sql_time2
data2 = jdt_db(sql)
```

```python
data = pd.concat([data1,data2],axis = 0)
```
